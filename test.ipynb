```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# 生成一个 3D Simplex
def generate_simplex():
    # 3D Simplex 的顶点是标准的 4 个点 (即 (0,0,0), (1,0,0), (0,1,0), (0,0,1))
    simplex_vertices = np.array([[0, 0, 0],
                                 [1, 0, 0],
                                 [0, 1, 0],
                                 [0, 0, 1]])
    return simplex_vertices

# 计算扰动后的 Simplex
def perturb_simplex(simplex_vertices, delta=0.05):
    # 将每个顶点在每个维度上随机扰动
    perturbed_simplex = simplex_vertices + np.random.uniform(-delta, delta, simplex_vertices.shape)
    return perturbed_simplex

# 绘制 3D Simplex
def plot_simplex(ax, simplex_vertices, color='b', alpha=0.5):
    # 连接 Simplex 的顶点
    verts = [
        [simplex_vertices[0], simplex_vertices[1], simplex_vertices[2]],
        [simplex_vertices[0], simplex_vertices[1], simplex_vertices[3]],
        [simplex_vertices[0], simplex_vertices[2], simplex_vertices[3]],
        [simplex_vertices[1], simplex_vertices[2], simplex_vertices[3]],
    ]

    # 绘制面
    ax.add_collection3d(Poly3DCollection(verts, facecolors=color, linewidths=1, edgecolors='k', alpha=alpha))

# 主函数
def main():
    # 创建 Simplex
    simplex_vertices = generate_simplex()

    # 创建扰动后的 Simplex
    perturbed_simplex_vertices = perturb_simplex(simplex_vertices, delta=0.05)

    # 创建 3D 图形
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # 绘制原始 Simplex
    plot_simplex(ax, simplex_vertices, color='blue', alpha=0.3)

    # 绘制扰动后的 Simplex
    plot_simplex(ax, perturbed_simplex_vertices, color='red', alpha=0.3)

    # 设置坐标轴
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()

if __name__ == "__main__":
    main()
```
